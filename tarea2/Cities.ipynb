```python
import pandas as pd

```

```python
cities = pd.read_csv('Pathfinder/data.csv')
```

```python
def dist(x1, y1, x2, y2):
    return ((x1-x2)**2 + (y1-y2)**2)**0.5
```

```python
distance = {}

for i in range(len(cities)):
    for j in range(len(cities)):
        if i not in distance:
            distance[i] = {}
        distance[i][j] = dist(cities['x'][i], cities['y'][i], cities['x'][j], cities['y'][j])
```

```python
top = sorted(distance[0].items(), key=lambda x:x[1])[:10]
for t in top:
    print(t[0])
```

```python
def trimDict(di, n):
    d = {}
    for i in range(len(di)):
        d[i] = sorted(distance[i].items(), key=lambda x:x[1])[1:n+1]
    return d
```

```python
def trimDict2(di, n):
    d = {}
    for i in range(len(di)):
        d[i] = {}
        top = sorted(distance[i].items(), key=lambda x:x[1])[1:n+1]
        for t in top:
            d[i][t[0]] = t[1]
    return d
```

```python
connections = trimDict(distance, 5)
```

```python
con2= trimDict2(distance, 5)
```

```python
sum([len(con2[x]) for x in con2])
```

```python
def fillDict(d):
    for k in d:
        for k2 in d[k]:
            d[k2][k] = d[k][k2]
        
    return d
                
```
